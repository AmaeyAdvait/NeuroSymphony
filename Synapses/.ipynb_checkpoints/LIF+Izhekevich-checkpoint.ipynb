```python
from brian2 import *

start_scope()
prefs.codegen.target = "numpy"
N = 30
```

```python
# LIF
tau = 10*ms
Vt = -50*mV
Vr = -65*mV
El = -65*mV
R = 10*Mohm
I1 = 2.5*nA

eqs_LIF = '''
        dV/dt = (-(V-El)  +R*I1)/tau : volt
    '''
G1 = NeuronGroup(N, model = eqs_LIF, threshold = 'V>Vt', reset = 'V = Vr', method = 'exact')
G1.V = El
```

```python
# Izhekevich
a = 0.02/ms
b = 0.2/ms
c = -65*mV
d = 5*mV/ms
I2 = 15*mV/ms

eqs_Izh = '''
         dv/dt = (0.04*v**2/ms/mV + 5*v/ms + 140*mV/ms - u + I2) : volt
         du/dt = a*(b*v-u) : volt/second
      '''  
G2 = NeuronGroup(N, model = eqs_Izh, threshold = 'v>=-30*mV', reset = 'v=c; u += d', method = 'euler')
G2.v = c
G2.u = b*G2.v
```

```python
def visualise_connectivity(S):
    Ns = len(S.source)
    Nt = len(S.target)
    figure(figsize=(10, 4))
    subplot(121)
    plot(zeros(Ns), arange(Ns), 'ok', ms=10)
    plot(ones(Nt), arange(Nt), 'ok', ms=10)
    for i, j in zip(S.i, S.j):
        plot([0, 1], [i, j], '-k')
    xticks([0, 1], ['Source', 'Target'])
    ylabel('Neuron index')
    xlim(-0.1, 1.1)
    ylim(-1, max(Ns, Nt))
    subplot(122)
    plot(S.i, S.j, 'ok')
    xlim(-1, Ns)
    ylim(-1, Nt)
    xlabel('Source neuron index')
    ylabel('Target neuron index')
```

```python
S = Synapses(G1, G2, on_pre = 'v_post += 0.02*mV')
S.connect(condition = 'i!= j', p = 0.2)
visualise_connectivity(S)
```

```python
M1 = StateMonitor(G1, 'V', record=True)
M2 = StateMonitor(G2, 'v', record=True)
run(100*ms)

plot(M1.t/ms, M1.V[0]/mV, label='LIF neuron')
plot(M2.t/ms, M2.v[0]/mV, label='Izhikevich neuron')
xlabel('Time (ms)')
ylabel('Membrane potential (mV)')
legend()
```

```python
spikemon_LIF = SpikeMonitor(G1)
spikemon_Izh = SpikeMonitor(G2)

run(200*ms)
figure(figsize = (6,4))
plot(spikemon_LIF.t/ms, spikemon_LIF.i, 'b.', label = 'LIF Neuron')
plot(spikemon_Izh.t/ms, spikemon_Izh.i, 'r.', label = 'Izhekevich Neuron')
xlabel('Time(ms)')
ylabel('Neuron index')
legend()
```
