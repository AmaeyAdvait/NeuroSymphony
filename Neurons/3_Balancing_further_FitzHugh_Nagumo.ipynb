```python
from brian2 import *

start_scope()
```

```python
a = 0.7
b = 0.8
ep = 0.08
I = 0.5

eqs = '''
        dv/dt = (v - v**3/3 - w + I)/ms : 1
        dw/dt = (ep*(v + a - b*w))/ms : 1
      '''

G = NeuronGroup(1, model = eqs, method = 'euler')
G.v = -1
G.w = 1

M = StateMonitor(G, ['v', 'w'], record = True)
defaultclock.dt = 0.01*ms
run(200*ms)
```

```python
figure(figsize = (10,4))
plot(M.t, M.v[0], label = 'Voltage')
plot(M.t, M.w[0], label = 'Recovery')
xlabel('Time')
ylabel('Variables')
title('FitzHugh Nagumo Neuron')
legend()
```
